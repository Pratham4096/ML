```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Automobile.csv")
df.head()
```

```python
plt.plot(df['horsepower'],marker='o',color='blue')

plt.title("Horsepower Line Chart")
plt.xlabel("Car Index")
plt.ylabel("Horsepower")
plt.grid(True)

plt.show()
```

```python
correlation = df.corr(numeric_only=True)

plt.figure(figsize=(8,6))

plt.imshow(correlation,
           cmap='coolwarm')

plt.colorbar()

plt.xticks(range(len(correlation.columns)),
           correlation.columns,
           rotation=90)

plt.yticks(range(len(correlation.columns)),
           correlation.columns)

plt.title("Correlation Heatmap")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["mpg"],
         bins=10,
         color='green',
         edgecolor='black')

plt.title("Histogram of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,6))

scatter = plt.scatter(
    df["weight"],
    df["horsepower"],
    c=df["mpg"],
    cmap='viridis'
)

plt.colorbar(scatter, label="MPG")

plt.title("Weight vs Horsepower")
plt.xlabel("Weight")
plt.ylabel("Horsepower")

plt.show()
```

```python
origin_count = df["origin"].value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    origin_count,
    labels=origin_count.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Cars by Origin")

plt.show()
```
