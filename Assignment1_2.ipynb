```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("/content/HousingData.csv")

print(df.describe())
```

```python
plt.figure(figsize=(6,6))

plt.boxplot(df["MEDV"])

plt.title("Boxplot of MEDV")
plt.ylabel("Median Value of Owner Occupied Homes")

plt.show()
```

```python
plt.hist(df['CHAS'])
plt.title("CHAS Histogram")
plt.xlabel("CHAS")
plt.ylabel("Frequency")
```

```python
plt.scatter(df["B"],df["NOX"])
plt.xlabel("Bussiness")
plt.ylabel("Nitric Oxide")
plt.title("Scatter Plot of Business and Nitric Oxide")
```

```python
plt.hist(df['PTRATIO'])
plt.title("TEACHER TO PUPIL RATIO")
plt.xlabel("TEACHER")
plt.ylabel("PUPIL")
```
